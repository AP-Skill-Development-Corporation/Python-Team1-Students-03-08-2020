## Matplotlib
- It is a cross-platform data visualization library for making 2D plots from data

### Types of plotting
- Line plot
- Bar Graph
- Scatter plot
- Area Plot
- Pie Charts
- Box Plot

```python
import matplotlib.pyplot as plt
```

### Line plot

```python
plt.plot([1,2,3,4],[2,4,6,8])
plt.show()
```

```python
plt.plot([1,2,3,4],[2,4,6,8],'g')
plt.show()
```

### label(),title(),legend()

```python
plt.plot([1,2,3,4],[2,4,6,8],'g')
plt.title('Line Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
import numpy as np
x = np.linspace(0,2,100)
x
```

```python
plt.plot(x,x,color='r',label='Normal Line')
plt.plot(x,x**2,color='g',label='Square Line')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plotting')
plt.legend()
plt.show()
```

### Bar Graph
- Used to create a Bar Chart

```python
plt.bar([1,2,3,4],[10,30,40,45],width=0.5)
plt.title('Bar Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axix')
plt.show()
```

```python
marks = [50,67,70,76,80]
sbj = ['C','C++','Java','Oracle','Python']
plt.bar(sbj,marks,color=['red','blue','green','pink','orange'])
plt.show()
```

### Scatter Plot

```python
plt.scatter([1,2,3,4],[5,6,7,8],color='b')
plt.show()
```

### Area Plot

```python
weeks = [1,2,3,4,5]
working = [5,7,4,2,6]
sleeping = [6,2,1,3,5]
plt.stackplot(weeks,working,sleeping,colors=['r','g'],labels=['Work','Rest'])
plt.legend()
plt.show()
```

### Box Plot

```python
marks = np.random.randint(20,100,101)
marks
```

```python
plt.boxplot(marks,patch_artist=True)
plt.show()
```

```python
v1 = np.random.randint(50,100,10)
v2 = np.random.randint(40,90,10)
v3 = np.random.randint(30,100,10)
print(v1)
print('.......')
print(v2)
print('........')
print(v3)
```

```python
plt.boxplot([v1,v2,v3],labels=['val1','val2','val3'])
plt.show()
```

### Pie Chart

```python
marks = [20,18,70,76,80]
sbj = ['C','C++','Java','Oracle','Python']
plt.pie(marks, labels=sbj)
plt.legend()
plt.show()
```
