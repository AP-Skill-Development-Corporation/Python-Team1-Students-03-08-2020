```python
# basic linear plot
import matplotlib.pyplot as plt
x=[25,26,27,28,29]
y=[29000,30000,31000,32000,34000]
plt.plot(x,y,label='yaxis',color='#008000',linestyle='--',marker='o')
y1=[28000,32000,49000,50000,65000]
plt.plot(x,y1,label='y1axis',color='y',linestyle='-.',marker='d')
plt.tight_layout()
plt.title('basic linear plot')
plt.xlabel('ages')
plt.ylabel('salary')
plt.grid(True)
plt.legend()
plt.style.use('fivethirtyeight')
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

```python
import pandas as pd
m=pd.read_csv('mpg.csv')
m.head()
```

```python
# bar graph
import matplotlib.pyplot as plt
plt.bar(m['cty'],m['hwy'])
plt.tight_layout()
plt.show()
```

```python
# barh graph
import matplotlib.pyplot as plt
plt.barh(m['cty'],m['hwy'])
plt.tight_layout()
plt.show()
```
